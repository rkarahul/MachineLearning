# Using List Line_chart

```python
import matplotlib.pyplot as plt
```

```python
#line chart
over=[5,10,15,20]
runs=[50,70,125,240]
plt.plot(over,runs)
plt.xticks(range(0,22,2))
plt.yticks(range(0,250,20))
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]
plt.plot(years,sales)
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]
plt.plot(years,sales)
plt.xticks(range(2018,2025,1))
plt.yticks(range(500000,2000000,200000))
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=['10L','15L','6L','8L','12L','15L']
plt.plot(years,sales)
plt.xticks(range(2018,2025,1))
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]plt.plot(years,sales,c='b',linewidth=2,linestyle='solid',marker='o',markerfacecolor='r',markersize=10,markeredgecolor='g')

plt.xticks(range(2018,2025,1))
plt.yticks(range(500000,2000000,200000))
plt.xlabel('Years')
plt.ylabel('Sales')
plt.text(2017.8,1050000,1000000)
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]
plt.plot(years,sales,c='b',linewidth=2,linestyle='solid',marker='o',markerfacecolor='r',markersize=10,markeredgecolor='g')
plt.xticks(range(2018,2025,1),c='b',size=12)
plt.yticks(range(500000,2000000,200000),c='b',size=12)
plt.xlabel("Years",size=14)
plt.ylabel("Sales",size=14)
plt.title("B/S over time")
for y,s in zip(years,sales):
    plt.text(y-.2,s+50000,s,c='g',size=10)
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales_1=[1000000,1500000,600000,800000,1200000,1500000]
sales_2=[900000,1500000,1600000,1800000,2200000,2500000]
plt.plot(years,sales_1,c='b',linewidth=2,linestyle='solid')
plt.plot(years,sales_2,c='k',linewidth=2,linestyle='solid')
plt.xticks(range(2018,2025,1),c='b',size=12)
plt.yticks(range(500000,3000000,200000),c='b',size=12)
plt.xlabel("Years",size=14)
plt.ylabel("Sales",size=14)
plt.title("B/S over time")
plt.legend(["ABC","XYZ"])
plt.grid()
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales_1=[1000000,1500000,600000,800000,1200000,1500000]
sales_2=[900000,1500000,1600000,1800000,2200000,2500000]
plt.plot(years,sales_1,c='b',linewidth=2,linestyle='solid')
plt.plot(years,sales_2,c='k',linewidth=2,linestyle='solid')
plt.xticks(range(2018,2025,1),c='b',size=12)
plt.yticks(range(500000,3000000,200000),c='b',size=12)
plt.xlabel("Years",size=14)
plt.ylabel("Sales",size=14)
plt.title("B/S over time")
plt.legend(["ABC","XYZ"])
plt.grid()
plt.savefig("E:/bis_com.png")
```

# Using line chart with Read_csv

```python
import pandas as pd
```

```python
df = pd.read_excel("E:\\DUCAT_PYTHON COURSE\\Sample - Superstore.xlsx")
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['year']=df['Order Date'].dt.year
df
```

```python
res_df=df.groupby(by="year")[['Sales']].sum()
res_df
```

```python
res_df.index
```

```python
res_df.Sales.values
```

```python
years=res_df.index
sales=res_df.Sales.values

plt.figure(figsize=(8,6))
plt.plot(years,sales,c='b',linewidth=2,linestyle='solid',marker='o',markerfacecolor='r',markersize=10,markeredgecolor='g')

plt.xlabel("Years",size=14)
plt.ylabel("Sales",size=14)
plt.title("B/S over time")
for y,s in zip(years,sales):
    plt.text(y-.2,s,s,c='g',size=12)
plt.show()
```

```python
years=res_df.index
sales=res_df.Sales.values

plt.figure(figsize=(8,6))
plt.plot(years,sales,c='b',linewidth=2,linestyle='solid',marker='o',markerfacecolor='r',markersize=10,markeredgecolor='g')

plt.xlabel("Years",size=14)
plt.ylabel("Sales",size=14)
plt.title("B/S over time")
plt.xticks(range(2020,2024,1))
plt.show()
```

```python
import pandas as pd
df = pd.read_excel("E:\\DUCAT_PYTHON COURSE\\Sample - Superstore.xlsx")
df['year']=df['Order Date'].dt.year
res_df=df.groupby(by="year")[['Sales']].sum()

years=res_df.index
sales=res_df.Sales.values

plt.figure(figsize=(8,6))
plt.plot(years,sales,c='b',linewidth=2,linestyle='solid',marker='o',markerfacecolor='r',markersize=10,markeredgecolor='g')

plt.xlabel("Years",size=14)
plt.ylabel("Sales",size=14)
plt.title("B/S over time")
plt.xticks(range(2020,2024,1))
plt.show()
```
